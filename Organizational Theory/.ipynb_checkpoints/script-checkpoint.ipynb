```python
import pandas as pd
import numpy as np
import re

#Enron_full = pd.read_csv("C:/Users/chris/Desktop/Epitech/Organizational Theory/Enron.csv")
users=pd.read_csv("C:/Users/chris/Desktop/Epitech/Organizational Theory/users.csv", header=None)
exchanges = pd.read_csv("C:/Users/chris/Desktop/Epitech/Organizational Theory/exchanges.csv", header=None)
```

```python
exchanges.head()
```

```python
users.head()
```

```python
full=exchanges.merge(users, how="left", left_on=1, right_on=0)
full.head()
full = full[["1_y",2,3,4]]
full.head()
full=full.merge(users, how="left", left_on=2, right_on=0)
full = full[["1_y", 1, 3, 4]]
full = full.rename({"1_y":"sender", 1:"receiver",3:"date",4:"mail_subject"}, axis='columns')
#full.columns = [["sender","receiver","date","mail_subject"]]

full["year"] = full["date"].str.split('-')
full.loc[:, 'year'] = full.date.map(lambda x: x[0:4])
enron_2010 = full[full["year"]=="2000"]
enron_2010.head()
```

```python
dot = {}
for index,row in enron_2010.iterrows():
    receiver = row["receiver"]
    sender = '"'+row["sender"]+'"'
    if receiver not in dot.keys():
        dot[receiver] = [sender]
    elif receiver in dot.keys() and sender not in dot[receiver]:
        dot[receiver].append(sender)

```

```python
result = ""
for k,v in dot.items():
    key = '"'+k+'"'
    value = " ".join(v)
    result =  result + key+ " -> { "+value+ " };"
    

final = "digraph { graph [ overlap =scale , splines =true , rankdir =LR ];" +result+ "}"
print(final)
```

```python
f= open("C:/Users/chris/Desktop/Epitech/Organizational Theory/script.dot","w+")
f.write(final)
f.close()
```
